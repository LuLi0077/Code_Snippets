```python
from ggplot import *
%matplotlib inline
```

```python
ggplot(meat, aes(x='date', ymin='beef - 1000', ymax='beef + 1000')) + \
    geom_ribbon() + \
    scale_x_date(labels=date_format('%B %-d, %Y'))
```

```python
ggplot(meat, aes(x='date', ymin='beef*0.9', ymax='beef*1.1')) + \
    geom_ribbon() + \
    scale_x_date(labels=date_format('%B %-d, %Y'))
```

```python
ggplot(meat, aes(x='date', y='beef', ymin='beef*0.9', ymax='beef*1.1')) + \
    geom_ribbon() + \
    geom_point() + \
    scale_x_date(labels=date_format('%B %-d, %Y'))
```
