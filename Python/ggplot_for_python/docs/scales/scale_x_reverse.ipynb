```python
%matplotlib inline
from ggplot import *
```

### `scale_x_reverse`
`scale_x_reverse` reverses your x-axis. Its parameters are:

```python
ggplot(diamonds, aes(x='price', y='carat')) + \
    geom_point() + \
    scale_x_reverse()
```

```python
ggplot(pigeons, aes(x='speed')) + \
    geom_histogram() + \
    scale_x_reverse()
```

```python
ggplot(pigeons, aes(x='speed', y='pos')) + \
    geom_line() + \
    scale_x_reverse()
```
